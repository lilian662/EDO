```python

import numpy as np
import matplotlib.pyplot as plt

# Definimos la funcin de perturbacin f(x') = exp(-((x' - 0.5)^2)/0.01)
def f(xp):
    return np.exp(-((xp - 0.5)**2) / 0.01)

# Parmetros
c = 1       # velocidad de propagacin
t = 0.5     # tiempo
x = np.linspace(0, 1, 500)  # eje espacial

# (a) Onda viajando a la derecha: x' = x - ct
onda_derecha = f(x - c * t)

# (b) Onda viajando a la izquierda: x' = x + ct
onda_izquierda = f(x + c * t)

# (c) Superposicin de las ondas anteriores
superposicion = onda_derecha + onda_izquierda

# Grfica
plt.figure(figsize=(10, 6))

# Graficar cada onda con estilos diferentes
plt.plot(x, onda_derecha, 'orange', linestyle='--', label="Onda a la derecha (x - ct)")
plt.plot(x, onda_izquierda, 'red', linestyle=':', label="Onda a la izquierda (x + ct)")
plt.plot(x, superposicion, 'deeppink', linewidth=2, label="Superposicin")

# Personalizacin de la grfica
plt.title("Ondas viajeras y su superposicin")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Mostrar
plt.show()
```
