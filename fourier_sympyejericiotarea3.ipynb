```python
from sympy import fourier_series, pi, plot, Piecewise,exp
from sympy.abc import x
```

```python
f=Piecewise(

    (1,(x >= 0) & (x <= pi/2)),
    (2,(x > pi/2) & (x <= pi)))
```

```python
plot(f, (x,0,pi),line_color ='red')
```

```python
s = fourier_series(f, (x, 0, pi))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, 0, pi), show=False, legend=True)
```

```python
s
```

```python
s3
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```
