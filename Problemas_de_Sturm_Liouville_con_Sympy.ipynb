```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de n
n_vals = [1, 2, 3]
x_vals = np.linspace(0, 1, 400)

# Crear figura
plt.figure(figsize=(10, 5))

# Graficar para cada valor de n
for n in n_vals:
    k = (n * np.pi) ** 2
    X = lambda x: np.sin(n * np.pi * x)  # Solucin con C=1
    y_vals = X(x_vals)

    plt.plot(x_vals, y_vals, label=f"n = {n}")

# Estilo
plt.title(r"Soluciones de $X'' + kX = 0$ con $X(0)=0$, $X(1)=0$")
plt.xlabel("x")
plt.ylabel("X(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
n_vals = [1, 2, 3]  # Valores de n
x_vals = np.linspace(0, 1, 400)

# Crear figura
plt.figure(figsize=(10, 5))

# Recorrer valores de n
for n in n_vals:
    k = (n * np.pi) ** 2
    sqrt_k = np.sqrt(k)

    # Condiciones: X'(0)=0  C2 = 0, as que usamos C1 = 1 arbitrariamente
    C1_val = 1
    C2_val = 0

    # Solucin: X(x) = C1 * cos(sqrt(k)*x)
    X = lambda x: C1_val * np.cos(sqrt_k * x)

    # Evaluar
    y_vals = X(x_vals)

    # Graficar
    plt.plot(x_vals, y_vals, label=f"n = {n}")

# Estilo
plt.title(r"Soluciones de $X'' + kX = 0$ con $X'(0)=0$, $X'(1)=0$")
plt.xlabel("x")
plt.ylabel("X(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
